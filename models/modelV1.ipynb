```python
import pandas as pd
```

```python
df = pd.read_csv('../data/convertData/finland_electricity_predict_dataset.csv')
df
```

# Create train & test data set

```python
#drop the target variable and the datetime column from the feature set
X = df.drop(columns = ['price', 'datetime'])

#set the dependent variable
y = df['price']
```

```python
from sklearn.model_selection import train_test_split

# spliting the data into training, testing and validation sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle=False)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#clean up the feature column names in the training and test sets, replacing square brackets with underscores or English words
X_train = X_train.rename(columns={'Air temperature mean [°C]': 'air_temp_mean', 'Wind speed mean [m/s]': 'wind_speed_mean'})
X_test = X_test.rename(columns={'Air temperature mean [°C]': 'air_temp_mean', 'Wind speed mean [m/s]': 'wind_speed_mean'})

#Print the latest cleaned feature list and perform a security check.
print("Purified Feature Names in Training Set:", X_train.columns.tolist())
```

# TRAIN V1 XG-BOOST REGRESSION MODEL 

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
```

```python
#instantiate the model with default hyperparameters
model_v1 = XGBRegressor(n_estimators=100, random_state=42)

print("Starting training V1 baseline model...")
#train the model on the training data
model_v1.fit(X_train, y_train)

print("Training finished! Starting testing...")
```

```python
# test the model on the test set and get predictions
y_pred = model_v1.predict(X_test)

# evaluate the model using mean absolute error (MAE)
mae_score = mean_absolute_error(y_test, y_pred)

print("\n======== V1 Model Evaluation ========")
print(f"V1 Baseline Model MAE: {mae_score:.2f} EUR/MWh")
```
