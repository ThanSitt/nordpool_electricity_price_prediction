```python
import pandas as pd

# API URL
url = (
   "https://sahkotin.fi/prices.csv"
    "?start=2022-12-31T22:00:00.000Z"
    "&end=2025-12-31T22:00:00.000Z"
)

# Download CSV directly into pandas
df = pd.read_csv(url)

# Convert UTC → Helsinki time
df['hour'] = pd.to_datetime(df['hour'], utc=True)
df['hour'] = df['hour'].dt.tz_convert('Europe/Helsinki')
print(df)
```

```python
print(df.head(5000))
df['hour'] = df['hour'].dt.strftime('%Y-%m-%d %H:%M:%S')

print(df.head(5000))
```
