```python
files = glob.glob(
    "e:/Github/nordpool_electricity_price_prediction/data/originalData/WindDirection&Speed/*.csv"
)

print(files)
```

```python
df_list=[]

for file in files:

    wind=pd.read_csv(file)

    df_list.append(wind)

wind_df=pd.concat(df_list)

print(wind_df.shape)
```

```python
wind_df.info();wind_df.describe();wind_df.sample(5)
```

```python
wind_df.isnull().sum(); 
```

```python
#convert the 'Wind speed mean [m/s]' column to numeric, and coerce any errors to NaN
wind_df['Wind speed mean [m/s]'] = pd.to_numeric(wind_df['Wind speed mean [m/s]'], errors='coerce')
wind_df['Wind direction mean [°]'] = pd.to_numeric(wind_df['Wind direction mean [°]'], errors='coerce')

#forward fill the missing values in the 'Wind speed mean [m/s]' column
wind_df['Wind speed mean [m/s]'] = wind_df['Wind speed mean [m/s]'].ffill()
#forward fill the missing values in the 'Wind direction mean [°]' column
wind_df['Wind direction mean [°]'] = wind_df['Wind direction mean [°]'].ffill()
```

```python
wind_df.info()
```

```python
#create a new column 'datetime_str' by concatenating the 'Year', 'Month', 'Day', and 'Time [Local time]' columns, and then convert this new column to a datetime object in a new column called 'datetime'
wind_df['datetime_str'] = wind_df['Year'].astype(str) + '-' + wind_df['Month'].astype(str) + '-' + wind_df['Day'].astype(str) + ' ' + wind_df['Time [Local time]']

#convert the 'datetime_str' column to a datetime object in a new column called 'datetime'
wind_df['datetime'] = pd.to_datetime(wind_df['datetime_str'])
```

```python
#set the 'datetime' column as the index of the winddataframe
wind_df.set_index('datetime', inplace=True)

hourly_wind_speed = wind_df['Wind speed mean [m/s]'].resample('h').mean().reset_index()
```
