```python
import pandas as pd
# import numpy as np
# from pathlib import Path
# import seaborn as sns
# import glob
# import matplotlib.pyplot as plt
```

```python
# read data
df_price = pd.read_csv("../originalData/electricPrices/electricity_prices_helsinki.csv")
df_temp = pd.read_csv("../originalData/Temperature/Temperature_hourly_clean.csv")
df_wind = pd.read_csv("../originalData/WindDirection&Speed/hourly_wind_speed.csv")

# change the time column from string to datetime type
df_price['hour'] = pd.to_datetime(df_price['hour'])
df_temp['datetime'] = pd.to_datetime(df_temp['datetime'])
df_wind['datetime'] = pd.to_datetime(df_wind['datetime'])
```

```python
#first merge: merge price and temp tables together
merge_1 = pd.merge(df_price, df_temp, left_on='hour', right_on='datetime', how='inner')

#second merge: merge the just-merged table with the wind table
final_df = pd.merge(merge_1, df_wind, left_on='hour', right_on='datetime', how='inner')
```

```python
final_df.head()
```

```python
# delete the redundant time columns
final_df_clean = final_df.drop(columns=['datetime_x', 'datetime_y'])

# rename the 'hour' column to 'datetime' for better readability
final_df_clean = final_df_clean.rename(columns={'hour': 'datetime'})

# save the cleaned and merged dataset to a new CSV file
final_df_clean.to_csv("../convertData/finland_electricity_predict_dataset.csv", index=False)
```
